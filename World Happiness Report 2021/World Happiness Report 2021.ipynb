```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE5B4'
```

```python
report_df = pd.read_csv('world-happiness-report-2021.csv')
```

```python
report_df
```

```python
report_df.info()
```

```python
report_df.describe().transpose()
```

```python
data_coloumns = ['Country name','Regional indicator','Happiness score','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy',
                'Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
data = report_df[data_coloumns].copy()
```

```python
data.head()
```

```python
data.rename(columns={'Country name':'country_name','Regional indicator':'regional_indicator','Ladder score':'ladder_score','Logged GDP per capita':'logged_gdp_per_capita','Social support':'social_support','Healthy life expectancy':'healthy_life_expectancy', 'Freedom to make life choices':'freedom_to_make_life_choices','Generosity':'generosity','Perceptions of corruption':'perception_of_corruption','Happiness score':'happiness_score' },inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
gdp_region = data.groupby('regional_indicator')['logged_gdp_per_capita'].sum()
gdp_region
```

```python
plt.figure(figsize=(10,7),dpi=70)
plt.pie(gdp_region,autopct='%1.1f%%')
plt.ylabel('');
```

```python
plt.figure(figsize=(10,7),dpi=100)
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP per region')
plt.ylabel(' ');
```

```python
# total countries

total_countries= data.groupby('regional_indicator')[['country_name']].count()
total_countries
```

```python
# correlation map

plt.figure(figsize=(10,7),dpi=100)
cor = data.corr(method = 'pearson')
sns.heatmap(cor,cmap='Blues',annot = True);
```

```python
# corruption in regions
corruption = data.groupby('regional_indicator')['perception_of_corruption'].mean()
corruption
```

```python
plt.figure(figsize=(12,8),dpi=100)
plt.title('Perceptions of Corruption in Verious Regions')
plt.xticks(rotation = 30,ha = 'right')
plt.xlabel('corruption index')
plt.ylabel('Region')
sns.barplot(corruption.index,corruption)
```

```python
top_10 = data.head(10)
bottom_10 =data.tail(10)
```

```python
bottom_10
```

```python
top_10.country_name
```

```python
fig,axes = plt.subplots(1,2,figsize = (16,6))
# plt.tight_layout(pad=2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 happiest countries life expentancy')
axes[0].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=xlabels,y=top_10.healthy_life_expectancy,ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('life expectancy')

ylabels= bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries life expentancy')
axes[1].set_xticklabels(ylabels,rotation=45,ha='right')
sns.barplot(x=ylabels,y=bottom_10.healthy_life_expectancy,ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('life expectancy');

```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.scatterplot(x=data.freedom_to_make_life_choices,y=data.happiness_score,hue= data.regional_indicator,s=200)
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score');
```

```python
country = data.sort_values('perception_of_corruption').head(10)
```

```python
plt.figure(figsize=(12,6),dpi=100)
plt.title('Countries with Most Perception of corruption')
plt.xticks(rotation=30,ha='right')
plt.bar(country.country_name,country.perception_of_corruption)
plt.ylabel('corruption index')

plt.xlabel('Countries Name')
```

```python
# corruption vs happiness
plt.figure(figsize=(12,6),dpi=100)
sns.scatterplot(data.happiness_score,data.perception_of_corruption,hue=data.regional_indicator,s=200)
plt.xlabel('Happiness')
plt.ylabel('Corruption');
```
