```python
def split_multicolumn(col_series):
    result_df=col_series.to_frame()
    options=[]
    #Iterate over the column
    for idx,value in col_series[col_series.notnull()].iteritems():
        #Break each value into list of options
        for option in value.split(';'):
            #Add the option asacolumn to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option]=False
            #Mark the value in the option column as True
            result_df.at[idx,option]=True
    return result_df[options]
                                 
```

```python
dev_type_df = split_multicolumn(survey.DevType)
```

```python
dev_type_df
```

```python
dev_type_total = dev_type_df.sum()#.sort_values(ascending=False)
dev_type_total
```

### Q. Which  were the most popular programming languages in 2019.

```python
schema.LanguageWorkedWith
```

```python
schema.LanguageDesireNextYear
```

```python
survey.LanguageWorkedWith.unique()
```

```python
survey.LanguageWorkedWith
```

```python
languages= split_multicolumn(survey.LanguageWorkedWith)
```

```python
languages
```

```python
popular_language = languages.sum().sort_values(ascending = False)
```

```python
popular_language
```

```python
# plt.xticks(rotation=85)
plt.figure(figsize=(12,6),dpi=100)
plt.title('popular languages')
sns.barplot(popular_language,popular_language.index)
plt.xlabel('count')
plt.ylabel('language');
```

```python

plt.figure(figsize=(12,6),dpi=100)
plt.title('popular languages')
plt.pie(popular_language,autopct='%1.1f%%',labels=popular_language.index);
# plt.legend(popular_language.index)
```

### Ans. popular language of 2019 is JavaScript.
